```python
import requests
import pandas as pd
import os
import json
print(requests.__version__)
print(pd.__version__)
print(json.__version__)
```

```python
url = 'https://api.github.com/search/repositories?q=language:python&sort=stars&order=desc&per_page=5'
response  = requests.get(url)
```

```python
if response.status_code == 200:
    data = response.json()['items']
    print("Request sent successfully!")
else:
    print("Request failed: %s" % response.status_code) 
print(len(data))
```

```python
df = pd.DataFrame(data)
df = df[['id', 'full_name', 'html_url', 'description', 'language', 'stargazers_count', 'forks_count', 'visibility', 'open_issues']]
owners_id = []
owners = []
for item in data:
    ow = item['owner']
    owners.append(ow)
    owners_id.append(ow['id'])
df['owner_id'] = owners_id
owners_df = pd.DataFrame(owners)
df.head()
```

```python
path = os.getcwd()
for item in data:
    dir_path = os.path.join(path, item['name'])
    if (not os.path.exists(dir_path)):
        os.mkdir(dir_path)
        file_path = dir_path + '/' + item['name'] + '.json'
    with open(file_path, 'w') as f:
        json.dump(item, f, indent=4)
    branchs = ['main', 'master']
    file_formats = ['README.md', 'README.rst']
    for branch in branchs:
        for readme in file_formats:
            url = 'https://raw.githubusercontent.com/'+item['full_name']+'/'+branch+'/'+ readme 
            res = requests.get(url)
            if res.status_code == 200:
                file_path = dir_path + '/' + readme
                with open(file_path, 'w') as f:
                    f.write(res.text)
                    break
```

```python
from collections import Counter
def word_count(file_name):
    with open(file_name) as f:
                return Counter(f.read().split())
```

```python
WORD_LIST = {}
for root, dirs, files in os.walk(path, topdown=False):
    for name in dirs:
        for readme in file_formats:
            readme_file = path + '/'+name+'/'+readme
            if (os.path.exists(readme_file)):
                words = word_count(readme_file)
                print(words)
                for word in words:
                    if word not in WORD_LIST:
                        WORD_LIST[word] = words[word]
                    else:
                        WORD_LIST[word] += words[word]
                break

WORD_LIST = sorted(WORD_LIST.items(), key=lambda x: -x[1])
WORD_LIST = pd.DataFrame(WORD_LIST[:100])
print("Most used words:")
(WORD_LIST)
```
