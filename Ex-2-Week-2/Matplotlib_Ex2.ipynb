1. Create a line plot with 50 data points that shows the relationship between two variables, x and y.

```python
import numpy as np
import matplotlib.pyplot as plt
# Create 50 points for x & y
x = np.linspace(0, 10, 50)
y = np.sin(x)
# Create the line plot
plt.plot(x, y)
# Add a tittle and labels to the plot
plt.title("Relationship between x and y")
plt.xlabel("x")
plt.ylabel("y")
# Show the plot
plt.show()
```

6. Use Matplotlib to create a stacked bar chart that compares the values of multiple
variables across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the categories and variables
categories = ['Apple', 'Samsung', 'Oppo', 'Huawei', 'Sony']
var1 = [20800, 18000, 10000, 15000, 12000]
var2 = [50000, 45000, 30000, 44000, 46000]
var3 = [100000, 90000, 65000, 78000, 85000]

# Create the stacked bar chart
fig, ax = plt.subplots()
ax.bar(categories, var1, label = 'Iphone')
ax.bar(categories, var2, bottom = var1, label = 'Tablets')
ax.bar(categories, var3, bottom = np.array(var1) + np.array(var2), label = 'Laptop')

# Add a title and labels to the chart
ax.set_title('Comparison of Multiple Variables Across Categories')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.legend()

# Show the chart
plt.show()
```

8. Create a heatmap that shows the relationship between two variables using a color
scale.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.rand(10, 10)
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = 'coolwarm')

cbar = ax.figure.colorbar(im, ax = ax)

ax.set_title("Heatmap of Two Variables")
ax.set_xlabel("Variable 1")
ax.set_ylabel("Variable 2")

plt.show()
```

9. Use Matplotlib to create a time series plot that shows the trend in a variable over time.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.arange('2010-01', '2015-01', dtype='datetime64[M]')
y = np.random.randint(0, 100, size=len(x))

# Create a time series plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Axis labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Variable')
ax.set_title('Trend in a Variable over Time')

# Show the plot
plt.show()
```

10. Create a 3D scatter plot that shows the relationship between three variables, x, y, and z.

```python
import numpy as np
import matplotlib.pyplot as plt
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c = z, cmap = 'coolwarm', alpha = 0.5)

# Axis labels and title
ax.set_xlabel("X")
ax.set_xlabel("Y")
ax.set_ylabel("Z")
ax.set_title("Relationship between X, Y and Z")

plt.show()
```
